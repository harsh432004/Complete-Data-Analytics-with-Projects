* Making Plot without line

```python
plt.plot(x,y,'o')
# 'o' -> Markers
```

```python
plt.plot(x,y,'d')
# 'd' ->Dimond Markers
```

```python
# References of lines

- -> Solid Line
: -> Dotted Line
-- -> Dashed Line
-. -> Dashed/Dotted Line
```

* Formating the String (fmt)

```python
SYNTAX: marker|line|color
```

```python
plt.plot(x,y,'d:y') # Here x,y are markers, d means dimond and y means yellow color
```

```python
# References of Colors
r -> Red
g -> Green
b -> Blue
y -> Yellow
w -> white
k -> Kala (means black)
```

```python
plt.plot(y,'d--k')
```

```python
plt.plot(y,marker='*',ms = 50) # ms means marker size
```

```python
plt.plot(y,marker='*',ms = 50, mec='r') # Here mec stands for marker edges color means star will have red edges
```

```python
plt.plot(y,marker='*',ms = 50, mec='#FF00FF',mfc='#9B7A7A') # mfc means marker face color means inside star color
# You can also have rgb colors. to know about rgb code of perticular color please visit: https://www.rapidtables.com/web/color/RGB_Color.html
```

* Grid -> adding grid to graph

```python
plt.plot(y)
plt.plot(x)

plt.xlabel('Height')
plt.ylabel('Weight')

plt.grid(color='black', ls='--', linewidth=2, axis='x')

plt.show()
```

* Subplots

```python
# 1st Plot
x = [1,2,3,4,5,6]
y = [10,20,30,40,50,60]

# plt.subplot(row, column, plot_numer)
plt.subplot(2,3,1) # this means row should be one, column should be two and plot should be in 1st plot
plt.plot(x,y)

# 2nd Plot
x = [3,8,1,10]
y = [6,2,7,11]
plt.subplot(2,3,2) # this means row should be two, column should be three and plot should be in 2nd plot
plt.plot(x,y)


# 3rd Plot
x = [1,2,3,4,5,6]
y = [10,20,30,40,50,60]

# plt.subplot(row, column, plot_numer)
plt.subplot(2,3,3)
plt.plot(x,y)

# 2nd Plot
x = [3,8,1,10]
y = [6,2,7,11]
plt.subplot(2,3,4)
plt.plot(x,y)


# 1st Plot
x = [1,2,3,4,5,6]
y = [10,20,30,40,50,60]

# plt.subplot(row, column, plot_numer)
plt.subplot(2,3,5)
plt.plot(x,y)

# 2nd Plot
x = [3,8,1,10]
y = [6,2,7,11]
plt.subplot(2,3,6)
plt.plot(x,y)

plt.suptitle("ANIMALS") # Heading in plots
```

* Ticks  & Ticks Label

* Bar Plot

```python
import matplotlib.pyplot as plt

x = ["One", "Two", "Three", "Four"]
y = [6, 9, 1, 20]

plt.bar(x, y, color="#9B7A8B", width=0.1) # to make horizontal bar just write plt.barh instead of plt.bar.
# always remember that with bar only width will be used and with barh only height will be used
plt.legend(title="Ice Cream flavors")
```

* Pie

```python
x = [25,25,35,15] # here lets assume 25,25,35,15 are students liking diff flavours of ice-cream
```

```python
plt.pie(x,labels=["Chocolates","Vanila", "Strawberry", "Blackcurrent"],
        startangle=90,explode=[0.2,0,0.1,0],shadow=True, # startangle means the angle of starting here we have set it to 90deg. explode means it will move out slightly from the graph
        colors=["green", "yellow","black","blue"])
plt.legend(title="Ice Cream flavors")
```
