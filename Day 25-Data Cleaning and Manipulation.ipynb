# Quicksort
* Description: Quicksort is a divide-and-conquer algorithm. It works by selecting a 'pivot' element from the array and partitioning the other elements into two sub-arrays, according to whether they are less than or greater than the pivot.
* Time Complexity:
* Average: O(n log n)
* Worst: O(n^2) (though this is rare with good pivot selection)
* Space Complexity: O(log n) due to recursion stack
* Stability: Not stable (relative order of equal sort items might change)
* Usage: Good general-purpose sort that is fast on average.
# Mergesort
* Description: Mergesort is also a divide-and-conquer algorithm. It divides the array into two halves, recursively sorts them, and then merges the sorted halves.
* Time Complexity:
* Average and Worst: O(n log n)
* Space Complexity: O(n) due to auxiliary arrays used in merging
* Stability: Stable (relative order of equal sort items is preserved)
* Usage: Preferred when stability is required.
# Heapsort
* Description: Heapsort transforms the array into a heap, then repeatedly extracts the maximum element from the heap and rebuilds the heap until all elements are sorted.
* Time Complexity:
* Average and Worst: O(n log n)
* Space Complexity: O(1) (in-place sorting)
* Stability: Not stable
* Usage: Useful when constant space usage is needed.

```python
df.sort_values(by=['Apples', 'Grapes'],kind='mergesort')
# 1. Quicksort
# 2. Mergesort
# 3. Heapsort
```
