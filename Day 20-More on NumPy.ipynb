```python
arr2.ndim
```

```python
# hstack() & vstack()
# hstack() is used to horizontally stack arrays, meaning it concatenates the arrays along the second axis (columns) if they are 2-dimensional.
# vstack() is used to vertically stack arrays, meaning it concatenates the arrays along the first axis (rows) if they are 2-dimensional.
```

```python
np.vstack([arr1,arr2,arr1,arr1])
```

```python
np.hstack([arr1,arr2])
```

* Any() & All()
* Any means koi bhi element voh condition satisfy karta hona chahiye.
* All means sabhi element voh condition satisfy karte hona chahiye.

```python
newarr = np.array([2,4,6,13,11])
```

```python
newarr
```

```python
# Any
np.any(newarr%2==0)
```

```python
# All
np.all(newarr%2==0)
```
