* around()

```python
arr = np.array([3.5016, 0.9846]) #creates a NumPy array with two elements, 3.501 and 0.98
```

```python
np.around(arr,decimals=3) # around func rounds up the value if val>5 then next digit else if val<5 then same digit.
# [0,0]
# [0,1]
# [1,1]
# [3.501, 1.000]
# [3.501, 0.98]
# [3.500, 1.000]
```

```python
# Lets understand view with simple example:
import numpy as np

arr = np.array([3.501, 0.98])
y = arr.view()

print("Original array (arr):", arr)
print("View of the array (y):", y)

# Modify the view
y[0] = 4.0
print("\nAfter modifying y:")
print("Original array (arr):", arr)
print("View of the array (y):", y)
```
