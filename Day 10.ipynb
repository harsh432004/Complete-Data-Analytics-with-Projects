```python
class First:
    def pass_or_fail(self, marks):
        if marks > 75:
            print("A Grade")
        elif 75 >= marks > 60:
            print("B Grade")
        elif 60 >= marks > 40:
            print("C Grade")
        elif 40 >= marks > 32:
            print("D Grade")
        else:
            print("Fail")
```

```python
first_instance = First()
marks = int(input("Enter marks: "))
first_instance.pass_or_fail(marks)
```

# Single Inheritance.

```python
class Animal:
  def speak(self):
    print("Animal is speaking")

class Dog(Animal):
  def bark(self):
    print("The dog barks")
  def speak(self):
    print("Cat is speaking")
```

```python
obj = Dog()  # First make an object of class
```

```python
obj.bark() #Call the object by objectname.function
```

```python
obj.speak()

# Animal is speaking - Parent class will get preference
# Cat is speaking
```

# Multi Level

```python
class Animal:
  def speak(self):
    print("Animal is speaking")

class Dog(Animal):
  def bark(self):
    print("The dog barks")

class Puppy(Dog):
  def eat(self):
    print("The puppy is eating")
  def bark(self):
    print("The puppy is not barking")
  
```

# Method Overiding

```python
class Animal:
  def speak(self):
    return 10

class Dog(Animal):
  def speak(self):
    return 7

class Cat(Animal):
  def speak(self):
    return 2
```

```python
obj1 = Animal()
obj2 = Dog()
obj3 = Cat()
```

```python
obj2.speak()
```
